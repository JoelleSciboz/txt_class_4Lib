```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, KFold, cross_val_score, cross_val_predict, GridSearchCV
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
from sklearn import metrics
from sklearn.metrics import f1_score,precision_score, recall_score, classification_report
from sklearn.pipeline import Pipeline
```

```python
title_features = (pd.read_csv('../data/C_learning/sc_2000_20007_title_features.csv')
                 .set_index('record_id')
                 )
```

```python
ft_features = (pd.read_csv('../data/C_learning/sc_2000_20007_title_features.csv')
                 .set_index('record_id')
                 )
```

```python
binary_labels = (pd.read_csv('../data/C_learning/sc_2000_2017_binary_labels.csv')
           .set_index('record_id')
           )
```

```python
labelset = binary_labels.T
```

```python
labelset['count'] = labelset.sum(axis=1)
```

```python
labelset.sort_values('count', ascending=False)
```

```python
labelset = labelset[labelset['count'] >= 500]
labelset.drop('count', inplace=True, axis=1)
len(labelset)
```

```python
labelset = labelset.T
labelset['count'] = labelset.sum(axis=1)
labelset = labelset[labelset['count'] > 0]
labelset.drop('count', inplace=True, axis=1)
len(labelset)
```

```python
record_id = labelset.index.tolist()
features = title_features.copy()
features = ft_features[ft_features.index.isin(record_id)]
```

```python
len(features) == len(labelset)
```

```python
X = features
y = labelset
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
clf = OneVsRestClassifier(LinearSVC(class_weight='balanced' ))
```

```python
clf.fit(X_train, y_train)
```

```python
y_predict = clf.predict(X_test)
```

```python
y_proba = clf.predict_proba(X_test)
```
